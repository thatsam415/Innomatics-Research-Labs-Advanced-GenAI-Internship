## Problem 1: Employee Performance Bonus Eligibility

```python
# Employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

## Problem 2: Search Query Keyword Analysis

```python
import string

# Input search query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display only words with frequency > 1
result = {word: count for word, count in word_count.items() if count > 1}

print(result)
```

## Problem 3: Sensor Data Validation

```python
# Sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid readings (even numbers)
valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

```python
# List of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate and display percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

## Problem 5: Sales Spike Detection

```python
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Detect spikes (more than 30% above average)
threshold = average_sales * 1.3

for index, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {index}: {value}")
```

## Problem 6: Duplicate User ID Detection

```python
# User IDs list
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count user IDs
user_count = {}

for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicates only
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
