## Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Check viral condition
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example
likes = [200, 350, 500]
analyze_post_engagement(likes)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average_rainfall = total / len(rainfall_data)
    
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

# Example
rainfall = [70, 75, 80, 63]
rainfall_checker(rainfall, 70)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
def check_duplicate_accounts(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    print(f"Duplicate Accounts Found: {result}")

# Example
users = ["alice", "bob", "alice", "charlie"]
check_duplicate_accounts(users)
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python
def premium_crop_filter(prices):
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print(f"Premium Crops: {premium_crops}")

# Example
crop_prices = [1800, 2500, 3200, 1500]
premium_crop_filter(crop_prices)
```

## Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example
application_health_checker(7)
```

## Problem 8: Banking – Daily Transaction Limit Checker

```python
def transaction_limit_checker(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Example
transaction_limit_checker(60000)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    
    for day in attendance_list:
        if day == 1:
            attended_classes += 1
    
    attendance_percentage = (attended_classes / total_classes) * 100
    
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {status}")

# Example (1 = Present, 0 = Absent)
attendance = [1, 1, 1, 0, 1]
attendance_eligibility(attendance)
```

## Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill_calculator(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage

# Example
total_bill, usage_status = electricity_bill_calculator(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Status: {usage_status}")
```
